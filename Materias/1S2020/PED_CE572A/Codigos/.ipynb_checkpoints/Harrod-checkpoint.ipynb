# Simulando a instabilidade de Harrod

# Importando pacotes

```python
%config InlineBackend.figure_format = 'retina'

import pandas as pd
import matplotlib.pyplot as plt
import sympy as sp
import seaborn as sns
import matplotlib.animation as animation

from pysolve3.model import Model
from pysolve3.utils import SFCTable, ShockModel, SolveSFC, SummaryShock
```

```python
def Harrod(
    s=.8,
    mu=.8,
    v=2,
    alpha=.2,
    u0=.75,
    Y0 = 100, g0=0.02
):
    
    model = Model()
    
    ######### Variáveis em nível #############
    model.var('Y', desc='PIB', default=Y0)
    model.var('It', desc='Investimento', default=s*Y0)
    
    ######### Variáveis em taxa ###############
    model.var('g', desc='Taxa de crescimento do produto', default=g0)
    model.var('gI', desc='Taxa de crescimento do investimento', default=g0)
    model.var('gK', desc='Taxa de crescimento do estoque de capital', default=(s/v)*u0)
    model.var('gw', desc='Taxa de crescimento garantida', default = (s/v)*mu)
    model.var('u', desc='Grau de utilização efetivo', default=u0)
    
    ########## Parâmetros #####################
    model.param('mu', desc='Grau de utilização normal', default=mu)
    model.param('alpha', desc='Velocidade de ajustamento', default=alpha)
    model.param('s', desc='Propensão marginal a poupar', default=s)
    model.param('v', desc='Relação técnica capital produto', default=v)
    
    
    ######### Equações ################
    model.add('Y = It/s')
    model.add('It = (1+gI)*It(-1)')
    model.add('d(gI) = alpha*(u(-1) - mu)')
    model.add('gK = (s/v)*u')
    model.add('u = (u(-1)*(1+g)/(1+gK))')
    model.add('gw = (s/v)')
    model.add('g = gI')
    
    return model
```

```python
base = Harrod()
df = SolveSFC(model=base, time=10)

fig, ax = plt.subplots(1,2)

df["g"].plot(ax=ax[0], color='black', title="Taxa de crescimento")
ax[0].axhline(y=df['gw'][0], color='gray', ls='--')
ax[0].legend()

df["u"].plot(ax=ax[1], color='black', title="Grau de utilização")
ax[1].axhline(y=df['mu'][0], color='gray', ls='--')
ax[1].legend()

sns.despine()
plt.tight_layout()
plt.show()
```

```python
base = Harrod()
df = SolveSFC(model=base, time=10)

fig, ax = plt.subplots(1,2, dpi=300, figsize=(16,5))

def animate(i):
    data = df.iloc[:int(i+1)] #select data range

    data["g"].plot(ax=ax[0], color='black', title="Taxa de crescimento")
    ax[0].axhline(y=data['gw'][0], color='gray', ls='--', label="Taxa garantida")
    

    data["u"].plot(ax=ax[1], color='black', title="Grau de utilização")
    ax[1].axhline(y=data['mu'][0], color='gray', ls='--')

    sns.despine()
    plt.tight_layout()

    
Writer = animation.writers['ffmpeg']
writer = Writer(fps=1, metadata=dict(artist='gpetrini'), bitrate=1800, )
    
gif = animation.FuncAnimation(fig, animate, frames=17, repeat=True)
gif.save('Harrod.mp4', writer=writer)
```
