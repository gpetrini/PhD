# Plots

```python
def plot_figures(variable, file):
    fig, ax = plt.subplots(figsize=(8,5))
    bank1[variable].plot(ax=ax, ls="-", color='black', label="1.0", lw=2)
    bank2[variable].plot(ax=ax, ls="-", color='darkred', label="2.5", lw=2)
    
    ax.legend(title="Bank")
    sns.despine()
    plt.show()
    fig.savefig('./figs/' + file+'.svg', transparent=True, dpi=600,
            bbox_inches='tight', pad_inches=0)
```

```python
for i in range(len(bank1.columns)):
    plot_figures(variable=bank1.columns[i], file=f'12_{i+1}')
```
