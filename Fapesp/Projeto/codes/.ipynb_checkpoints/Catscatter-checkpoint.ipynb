# Loading packages

```python
%config InlineBackend.figure_format = 'retina'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
sns.set_context('talk')
```

# Creating dataframe

```python
df = pd.DataFrame(
    data = {
        "Taxa de juros hipotecária": ["Fixa", "Fixa", "Variável", "Fixa", "Variável", "Variável"],
        "Reembolso antecipado": ["Ambos", "Ambos", "Legislado", "Contratado", "Legislado", "Ambos"],
        "Retirada de capital próprio": ["Não permitido", "Não permitido", "Não permitido", "Permitido", "Não permitido", "Indisponível"],
        "Financiamento pelo mercado de capitais" : ["Baixo", "Baixo", "Médio", "Médio", "Médio", "Alto"]
    }, 
    dtype="category",
    index = ["França", "Alemanha", "Itália", "Holanda", "Portugal", "Espanha"], 
)

df["Hipotecarização (1970)"] = [33, 42, 29, 46, 16, 12]
df["Países"] = df.index
df = df.sort_values(by=['Hipotecarização (1970)'], ascending=False)
df
```

# Catscatter

```python
def plotter(df,x,y,ax,legend=False):
    
    plot = sns.scatterplot(
        data=df, 
        x=x, y=y,
        size="Hipotecarização (1970)", 
        #color="black", style="Países",
        hue="Países",
        legend=legend,
        y_jitter=True, x_jitter=True,
        ax =ax,
        sizes=(0,500)
)
        
    plt.box(False)
    return plot
```

```python
fig, ax = plt.subplots(2,2, figsize=(16,10))

sns.scatterplot(
    data=df, hue="Países", size="Hipotecarização (1970)", sizes=(0,500),legend=False,
    x="Retirada de capital próprio",
    y="Financiamento pelo mercado de capitais",
    ax=ax[0,0]
)
ax[0,0].set_ylabel("Financiamento pelo\nmercado de capitais"); ax[0,1].yaxis.set_ticks_position('left')

sns.scatterplot(
    data=df, hue="Países", size="Hipotecarização (1970)", sizes=(0,500),legend=False,
    x="Reembolso antecipado",
    y="Financiamento pelo mercado de capitais",
    ax=ax[0,1]
)
ax[0,1].set_ylabel(""); ax[0,1].yaxis.set_ticks_position('none')

sns.despine(left=True, bottom=True)
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(2*8,5), sharey=True)
plotter(df, x="Retirada de capital próprio", y="Financiamento pelo mercado de capitais", ax=ax[0])
plotter(df, x="Reembolso antecipado", y="Financiamento pelo mercado de capitais", ax=ax[1])
ax[0].set_ylabel("Financiamento pelo \nmercado de capitais")
ax[0].set_xlabel("")
sns.despine(left=True, bottom=True)
fig1 = fig

fig, ax = plt.subplots(1,2, figsize=(2*8,5), sharey=True)
plotter(df, x="Retirada de capital próprio", y="Taxa de juros hipotecária", ax=ax[0])
plotter(df, x="Reembolso antecipado", y="Taxa de juros hipotecária", ax=ax[1], legend="brief")
ax[0].set_ylabel("Taxa de juros hipotecária")
sns.despine(left=True, bottom=True)
fig2 = fig
```

```python
fig1.add_axes(fig2)
```
