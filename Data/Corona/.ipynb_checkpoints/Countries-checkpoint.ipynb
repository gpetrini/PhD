# Loading packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

# Usefull objects

```python
selected_countries = [
  "Germany", "Italy", "United Kingdom", "Brazil", "China"
]

url = "https://covid.ourworldindata.org/data/full_data.csv"
```

# Downloading data

```python
df = pd.read_csv(url)
df.set_index(['location', 'date'], inplace=True)
df = df.loc[selected_countries]
df.describe()
```

```python
for country in selected_countries:
    print(f"\nFor {country}, the data is:")
    print(df.loc[country].describe())
    print("="*60)
    print("\n")
```

```python
def generic_plots(
    variable="new_deaths"
):
    for country in selected_countries:
        fig, ax = plt.subplots(figsize=(8,5))
        df.loc[country, variable].plot(
            ax=ax,
            title = f"{variable} in {country}",
            color='black',
        )
        sns.despine()
        plt.show()
```

```python
generic_plots(variable="new_deaths")
generic_plots(variable="new_cases")
generic_plots(variable="total_deaths")
generic_plots(variable="total_cases")
```
